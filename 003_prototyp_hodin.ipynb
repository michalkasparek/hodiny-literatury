```python
import json
import time
from datetime import datetime
```

```python
with open("cifernik.json","r",encoding="utf-8") as cif:
    cifernik = json.loads(cif.read())
```

```python
casy_na_ciferniku = [casy for casy, texty in cifernik.items()]
```

```python
def najdi_cas():
    global casy_na_ciferniku
    hodina = datetime.now().hour
    minuta = datetime.now().minute
    cas = f'{hodina:02}:{minuta:02}'
    while cas not in casy_na_ciferniku:
        minuta = minuta - 1
        cas = f'{hodina:02}:{minuta:02}'
    return cas
```

```python
def ukaz_cas():
    return cifernik[najdi_cas()]
```

```python
while True:
    print(f"{datetime.now().hour:02}:{datetime.now().minute:02} {cifernik[najdi_cas()]}")
    time.sleep(1)
```
