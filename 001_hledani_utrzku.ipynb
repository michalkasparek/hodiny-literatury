```python
import os
import ebooklib
from ebooklib import epub
from bs4 import BeautifulSoup
```

```python
import re
```

```python
# List to store dictionaries of epub files
knihy = []

# Path to the subfolder containing epub files
subfolder_path = "./knihy/thinkpad"

# Function to extract text content from epub file
def extract_text_from_epub(epub_file_path):
    book = epub.read_epub(epub_file_path)
    text_content = ''
    for item in book.get_items():
        if item.get_type() == ebooklib.ITEM_DOCUMENT:
            soup = BeautifulSoup(item.get_body_content(), 'html.parser')
            text_content += soup.get_text()
    return text_content

# Get list of epub files in the subfolder
epub_files = [file for file in os.listdir(subfolder_path) if file.endswith('.epub')]

# Iterate through each epub file
for epub_file in epub_files:
    # Create dictionary to store information
    book_info = {}
    
    # Get author and title from metadata
    book = epub.read_epub(os.path.join(subfolder_path, epub_file))
    try:
        author = book.get_metadata('DC', 'creator')[0][0]
    except:
        author = "???"
    try:
        title = book.get_metadata('DC', 'title')[0][0]
    except:
        title = "???"
    
    # Get text content of the epub file
    text = extract_text_from_epub(os.path.join(subfolder_path, epub_file))
    
    # Add information to dictionary
    book_info['Autor'] = author
    book_info['Titul'] = title
    book_info['text'] = text
    
    # Append dictionary to list
    knihy.append(book_info)
```

```python
def search_re(data, phrase, long=False):
    
    delka = 300
    if long == True:
        delka = 1200
    
    result = []
    # pattern = re.compile(re.escape(phrase))  # make sure the phrase is treated as literal
    pattern = re.compile(phrase)
    
    for item in data:
        text = item.get('text', '')
        match = pattern.search(text)
        if match:  # this means if the pattern is found
            start_index = max(0, match.start() - delka)  
            end_index = min(len(text), match.end() + delka)  
            snippet = text[start_index:end_index]
            marked = snippet[:match.start()] + snippet[match.start():match.end()] + snippet[match.end():]
            result.append(f"{marked}\n{item.get('Autor', '')}: {item.get('Titul', '')}")
    return result
```

```python
def hezky(seznam):
    for s in seznam:
        posl_radek = s.splitlines()[-1]
        s = s.replace(posl_radek, "")
        print(re.sub("\s"," ",s).replace("  "," ") + '\n' + posl_radek + "\n")
```

```python
casy_vsechny = []
minuty = [i for i in range(0,60)]
hodiny = [i for i in range(0,24)]
for h in hodiny:
    for m in minuty:
        casy_vsechny.append(f"{h:02}:{m:02}")
        casy_vsechny.append(f"{h}:{m:02}")
        casy_vsechny.append(f"{h:02}.{m:02}")
        casy_vsechny.append(f"{h}.{m:02}")
        casy_vsechny.append(f"{h}. hodin\w")
        casy_vsechny.append(f"{h} hodin(, | a ){m} minut")
casy_vsechny = list(set(casy_vsechny))
print(len(casy_vsechny))
```

```python
hezky(search_re(knihy, r'minut před',long=False))
```

```python
hezky(search_re(knihy, r' minut\w{0,1} po ', long=False))
```

```python
hezky(search_re(knihy, r'[Tt]ři[ \s]čtvrt. na tř', long=True))
```

```python
hezky(search_re(knihy, r' minut ', long=False))
```

```python
hezky(search_re(knihy, r'[Čč]tvrt hodiny před', long=False))
```

```python
hezky(search_re(knihy, "deset minut",long=True))
```

```python
for c in casy_vsechny:
    hezky(search_re(knihy, c.replace(".","\.")))
```
