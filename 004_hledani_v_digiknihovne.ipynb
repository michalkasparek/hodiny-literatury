```python
url = ["https://www.digitalniknihovna.cz/mzk/search?q=%22","%22"]
```

```python
cisla = [i for i in range(5,60)]
```

```python
po = 'první jedné druhé třetí čtvrté páté šesté sedmé osmé deváté desáté jedenácté dvanácté třinácté čtrnácté patnácté šestnácté sedmnácté osmnácté devatenácté dvacáté jednadvacáté dvaadvacáté třiadvacáté'.split(" ")
```

```python
for p in po:
    for c in cisla:
        print(f"{url[0]}{c}%20minut%20po%20{p}%20hodině{url[1]}")
```

```python
po_cislo = [f"{i}" for i in range(1,24)]
```

```python
for p in po_cislo:
    for c in cisla:
        print(f"{url[0]}{c}%20minut%20po%20{p}.%20hodině{url[1]}")
```

```python
for p in po_cislo:
    for c in cisla:
        print(f"{url[0]}{p}%20hodin%20a%20{c}%20minut{url[1]}")
```
